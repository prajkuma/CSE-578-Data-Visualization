```python
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('daily-minimum-temperatures-in-me.csv',delimiter=',',skip_header=1)
attend = data[:,1]

mean = np.nanmean(attend)
std = np.nanstd(attend)

plt.plot([0,len(attend)], [mean, mean], 'g-')
plt.plot([0,len(attend)], [mean+std, mean+std], 'y-')
plt.plot([0,len(attend)], [mean-std, mean-std], 'y-')
plt.plot([0,len(attend)], [mean+2*std, mean+2*std], 'r-')
plt.plot([0,len(attend)], [mean-2*std, mean-2*std], 'r-')
plt.plot(range(len(attend)), attend, 'b-')
plt.show()
```

```python
window_size = 50
plt.plot(np.convolve(attend, np.ones(window_size,)/window_size, 'same'))
plt.show()
```

```python
import pandas as pd

span = 50
df = pd.DataFrame({'att':attend})
df_exp = df.ewm(span=50).mean()
plt.plot(df_exp)
plt.show()
```
