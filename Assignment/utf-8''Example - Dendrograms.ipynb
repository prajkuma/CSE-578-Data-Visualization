```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import sqlite3

# Set the filename for the databse.
db_filename = 'cereals.db'

# Connect to the database and create a cursor object.
conn = sqlite3.connect(db_filename)
c = conn.cursor()

# Execute an SQL query to get a list of cereal manufacturers
#  in the dataset.
c.execute("SELECT Protein, Fat, Carbohydrates FROM cereals")
results = sum(map(list, list(c.fetchall())), [])
X = np.matrix(results).reshape(-1,3)

links = linkage(X, 'centroid')

dendrogram(links)
plt.show()
```

```python
from scipy.cluster.vq import whiten

Y = whiten(X)

links = linkage(Y, 'centroid')

dendrogram(links)
plt.show()
```
